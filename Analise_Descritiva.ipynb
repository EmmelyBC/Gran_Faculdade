## üìö Importando bibliotecas

```python
# Bibliotecas de machine learning e fun√ß√µes estat√≠sticas
import pandas as pd
import numpy as np

# Bibliotecas para plotagem de gr√°ficos
import matplotlib.pyplot as plt
import seaborn as sns
```

## üìù Criando um conjunto de dados

```python
np.random.seed(42) #semente que armazena a aleatoriedade usada

# Algoritmo que traduz a organiza√ß√£o da data frame
data = {
    'Idade': np.random.randint(18, 65, 1000), #(menor valor, maior valor, quantidade de linhas)
    'Sal√°rio': np.random.randint(25000, 150000, 1000),
    'N√≠vel de Educa√ß√£o': np.random.choice(['Ensino M√©dio', 'Gradua√ß√£o', 'P√≥s-Gradua√ß√£o'], 1000),
    'Cidade': np.random.choice(['S√£o Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Bras√≠lia'], 1000),
}

# Transforma√ß√£o do algoritmo armazenado na vari√°vel data em um dataframe armazenado na vari√°vel df
df = pd.DataFrame(data)
```

## üìä Realizando an√°lise descritiva

```python
# Analizando as 10 primeiras linhas do data frame
print(df.head(10))
```

```python
# Observando Estat√≠sticas Descritivas B√°sicas
print(df.describe())
```

```python
# Criando uma visualiza√ß√£o para entender a frequ√™ncia dos dados a respeito das idades - Histograma

plt.figure(figsize=(8, 6)) # tamanho do gr√°fico
sns.histplot(df['Idade'], bins=20, kde=True) # varia√ß√£o das idades no eixo x e linha de tend√™ncia no gr√°fico
plt.title('Distribui√ß√£o de Idades') # T√≠tulo do gr√°fico
plt.xlabel('Idade') # T√≠tulo do eixo X
plt.ylabel('Frequ√™ncia') # T√≠tulo do eixo y
plt.show() # Exibi√ß√£o do gr√°fico criado
```

<p>√â poss√≠vel perceber atrav√©s deste gr√°fico histograma que as idades que tem mais frequ√™ncia no nosso DataFrame s√£o um pouco mais de 20, 40 e 60 anos. Por√©m a idade que tem a maior frequ√™ncia √© de um pouco mais de 40 anos.</p>
